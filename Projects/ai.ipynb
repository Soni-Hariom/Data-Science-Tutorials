```python
import random

# Gossip-related responses for the chatbot to use
gossip_responses = [
    "Oh my gosh, I can't believe that!",
    "Wow, that's juicy!",
    "I had no idea. Thanks for the scoop!",
    "That's so scandalous!"
]

# Greeting messages for the chatbot to use
greeting_messages = [
    "Hello! Do you want to hear some gossip?",
    "Hi there! Want to hear some juicy news?",
    "Hello, friend! Up for some gossip?"
]

# Farewell messages for the chatbot to use
farewell_messages = [
    "Thanks for chatting with me!",
    "Goodbye for now!",
    "It was great gossiping with you. See you later!"
]

def gossip_chatbot():
    # Greet the user
    print(random.choice(greeting_messages))

    # Get the user's response
    response = input()

    # Check if the user wants to hear gossip
    if "yes" in response.lower():
        # Share a piece of gossip
        print("I heard that Kim Kardashian is secretly dating Kanye West's cousin!")

        # Ask the user if they have any gossip to share
        print("Do you have any gossip to share with me?")
        response = input()

        # If the user has gossip, exchange gossip
        while "yes" in response.lower():
            # Get the user's gossip
            print("What do you know?")
            user_gossip = input()

            # Respond with a gossip-related response
            print(random.choice(gossip_responses))

            # Ask the user if they have more gossip to share
            print("Do you have any more gossip to share?")
            response = input()

        # If the user does not have any more gossip, ask for their thoughts on the gossip shared
        print("What do you think about the gossip I shared with you earlier?")
        response = input()
        print("Interesting perspective. Thanks for sharing your thoughts with me.")

    # If the user does not want to hear gossip, end the conversation
    else:
        print(random.choice(farewell_messages))

# Start the gossip chatbot
gossip_chatbot()
```

```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
# Gossip-related responses for the chatbot to use
gossip_responses = [
    "Oh my gosh, I can't believe that!",
    "Wow, that's juicy!",
    "I had no idea. Thanks for the scoop!",
    "That's so scandalous!"
]

# Greeting messages for the chatbot to use
greeting_messages = [
    "Hello! Do you want to hear some gossip?",
    "Hi there! Want to hear some juicy news?",
    "Hello, friend! Up for some gossip?"
]

# Farewell messages for the chatbot to use
farewell_messages = [
    "Thanks for chatting with me!",
    "Goodbye for now!",
    "It was great gossiping with you. See you later!"
]

def classify_gossip(text):
    """
    Use a Naive Bayes classifier to determine if the input text is gossip-related or not.
    """
    # Create a list of tuples, where each tuple consists of a list of words and a label
    documents = []
    for category in movie_reviews.categories():
        for fileid in movie_reviews.fileids(category):
            documents.append((list(movie_reviews.words(fileid)), category))

    # Shuffle the documents
    random.shuffle(documents)

    # Create a list of all words in the movie reviews corpus
    all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
    word_features = list(all_words.keys())[:2000]

    # Define a feature extractor that will return a dictionary of the presence or absence of the
    # top 2000 most common words in the movie reviews corpus
    def document_features(document):
        document_words = set(document)
        features = {}
        for word in word_features:
            features['contains({})'.format(word)] = (word in document_words)
        return features

    # Extract features for all documents
    featuresets = [(document_features(d), c) for (d,c) in documents]

    # Split the feature sets into training and testing sets
    train_set, test_set = featuresets[100:], featuresets[:100]

    # Train a Naive Bayes classifier on the training set
    classifier = nltk.NaiveBayesClassifier.train(train_set)

    # Use the classifier to predict the class of the input text
    return classifier.classify(document_features(text.split()))

def gossip_chatbot():
    # Greet the user
    print(random.choice(greeting_messages))

    # Get the user's response
    response = input()

    # Check if the user wants to hear gossip
    if classify_gossip(response) == "pos":
        # Share a piece of gossip
        print("I heard that Kim Kardashian is secretly dating Kanye West's cousin!")

         # Ask the user if they have any gossip to share
        print("Do you have any gossip to share with me?")
        response = input()

        # If the user has gossip, exchange gossip
        while "yes" in response.lower():
            # Get the user's gossip
            print("What do you know?")
            user_gossip = input()

            # Respond with a gossip-related response
            print(random.choice(gossip_responses))

            # Ask the user if they have more gossip to share
            print("Do you have any more gossip to share?")
            response = input()

        # If the user does not have any more gossip, ask for their thoughts on the gossip shared
        print("What do you think about the gossip I shared with you earlier?")
        response = input()
        print("Interesting perspective. Thanks for sharing your thoughts with me.")

    # If the user does not want to hear gossip, end the conversation
    else:
        print(random.choice(farewell_messages))

# Start the gossip chatbot
gossip_chatbot()
```
